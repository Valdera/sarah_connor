```python
# show() or No show()? HOW TO DISPLAY YOUR PLOTS
# --- Plotting from a script ---
# plt.show() starts an event loop, looks for all currently active figure objects, and opens one or more interactive windows that display your figure or figures.
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```python
# --- Plotting from an IPython notebook ---
#  %matplotlib notebook will lead to interactive plots embedded within the notebook
#  %matplotlib inline will lead to static images of your plot embedded in the notebook
%matplotlib inline
import numpy as np
x = np.linspace(0, 10, 100)
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '-');
```

```python
# TWO INTERFACES FOR THE PRICE OF ONE
# --- MATLAB-style interface ---
plt.figure() # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

```python
# --- Object-oriented interface ---
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)
# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

```python
plt.plot(x, np.sin(x));
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

```python
# ADJUSTING THE PLOT: LINE COLORS AND STYLES
# The color can be specified in a variety of ways 
plt.plot(x, np.sin(x - 0), color='blue') # specify color by name
plt.plot(x, np.sin(x - 1), color='g') # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75') # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44') # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
# Similarly, you can adjust the line style using the linestyle keyword
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');
# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-') # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':'); # dotted
```

```python
# If you would like to be extremely terse, these linestyle and color codes can be combined into a single nonkeyword argument to the plt.plot()
plt.plot(x, x + 0, '-g') # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r'); # dotted red
```

```python
# ADJUSTING THE PLOT: AXES LIMIT
# The most basic way to adjust axis limits is to use the plt.xlim() and plt.ylim() methods
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

```python
plt.plot(x, np.sin(x))
plt.xlim(10, 0)
plt.ylim(1.2, -1.2);
```

```python
# passing a list that specifies [xmin, xmax, ymin, ymax]
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5])
```

```python
# The plt.axis() method goes even beyond this, allowing you to do things like automatically tighten the bounds around the current plot
plt.plot(x, np.sin(x))
plt.axis('tight');
```
